```python
import numpy as np
import glob
import os
from shutil import move
import matplotlib.pyplot as plt
from tqdm import tqdm
from PIL import Image
%matplotlib inline
```

Load the data path

```python
datapath = r'/media/banikr2/DATA/COVID19RadiographyDatabase'
dataloc = glob.glob(os.path.join(datapath,'*'))
```

```python
dataloc
```

## Dataset distribution

```python
covids = glob.glob(os.path.join(dataloc[0],'*.png'))
normals = glob.glob(os.path.join(dataloc[1],'*.png'))
pneumonias = glob.glob(os.path.join(dataloc[2],'*.png'))
COVID = len(covids)
Normal = len(normals)
Pneumonia = len(pneumonias)
sizes = [COVID, Normal, Pneumonia]
lesions = 'COVID', 'Normal', 'Pneumonia'
explode = (0.1, 0.01, 0.01)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=lesions, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### renaming the files in the folder

```python
for path in tqdm(pneumonias): #similar for COVID and normal names
    path_dest = path.replace('Viral Pneumonia (','')
    path_dest = path_dest.replace(')','')
    # print(path, path_dest)
    move(path,path_dest)
```
